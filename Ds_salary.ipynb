### Data Preprocessing

```python
df.info()
```

```python
df_preprocessed=df.copy()
df_preprocessed.drop(['salary','salary_currency'],axis=1,inplace=True)
```

```python
inputs=df_preprocessed.drop('salary_in_usd',axis=1)
targets=df_preprocessed['salary_in_usd']
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import make_column_transformer,make_column_selector
num_pipeline=make_pipeline(StandardScaler())
cat_pipeline=make_pipeline(OneHotEncoder())
preprocessing=make_column_transformer(
    (num_pipeline,make_column_selector(dtype_include=np.number)),
    (cat_pipeline,make_column_selector(dtype_include='category'))
)
```

```python
inputs_scaled=preprocessing.fit_transform(inputs)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(inputs_scaled,targets,test_size=0.2,random_state=42)
```

### Modelling

### 1. Linear Regression

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x_train,y_train)
```

```python
lin_pred=lin_reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
mse = mean_squared_error(y_test, lin_pred)
print("Mean Squared Error (MSE):", mse)
r2 = r2_score(y_test, lin_pred)
print("R Score:", r2)
mae = mean_absolute_error(y_test, lin_pred)
print("Mean Absolute Error (MAE):", mae)
```

```python
results_table=pd.DataFrame([['Linear Regression',mae,mse,r2 ]],columns=['Model','MAE','MSE','R']).round(3)
results_table
```

### 2. Elastic-Net Regression

```python
from sklearn.linear_model import ElasticNetCV
elas_net=model = ElasticNetCV(
    alphas=[0.1, 1.0, 10.0, 100.0],  
    l1_ratio=[0.1, 0.5, 0.7, 1.0],  
    cv=3,                           
    max_iter=10000,                  
    random_state=42
)
elas_net.fit(x_train,y_train)
elas_net_pred=elas_net.predict(x_test)
```

```python
print(elas_net.alpha_)
print(elas_net.l1_ratio_)
```

```python
mse = mean_squared_error(y_test, elas_net_pred)
print("Mean Squared Error (MSE):", mse)
r2 = r2_score(y_test, elas_net_pred)
print("R Score:", r2)
mae = mean_absolute_error(y_test, elas_net_pred)
print("Mean Absolute Error (MAE):", mae)
```

```python
results_table1=pd.DataFrame([['ElasticNet',mae,mse,r2]],columns=['Model','MAE','MSE','R']).round(3)
results_table=pd.concat([results_table,results_table1],ignore_index=True)
results_table
```

### 3. SGD regressor

```python
from sklearn.linear_model import SGDRegressor
from sklearn.model_selection import GridSearchCV
param_grid = {
    'alpha': [0.0001, 0.001, 0.01, 0.1],  
    'penalty': ['l2', 'l1', 'elasticnet'],  
    'l1_ratio': [0.15,0.4, 0.5, 0.7, 1.0],     # L1 ratio (only used if penalty='elasticnet')
}

sgd = GridSearchCV(SGDRegressor(), param_grid, cv=5,n_jobs=-1, scoring='neg_mean_squared_error' ) # Evaluation metric
sgd.fit(x_train,y_train)
sgd_pred=sgd.predict(x_test)
```

```python
sgd.best_params_
```

```python
mse = mean_squared_error(y_test, sgd_pred)
print("Mean Squared Error (MSE):", mse)
r2 = r2_score(y_test, sgd_pred)
print("R Score:", r2)
mae = mean_absolute_error(y_test, sgd_pred)
print("Mean Absolute Error (MAE):", mae)
```

```python
results_table1=pd.DataFrame([['SGD Regressor',mae,mse,r2]],columns=['Model','MAE','MSE','R']).round(3)
results_table=pd.concat([results_table,results_table1],ignore_index=True)
results_table
```

### 4. SVM Regressor

```python
from sklearn.svm import SVR
param_grid = {
    'C': [0.1, 1, 10, 100, 500, 1000, 2500,5000,10000],  # Regularization parameter
    'epsilon': [0.01, 0.1, 0.2, 0.5, 0.7, 0.9],     # Epsilon in the epsilon-SVR model
    'kernel': ['linear', 'poly', 'rbf'],   # Kernel type to be used in the algorithm
    'degree': [2, 3, 4]                    # Degree of the polynomial kernel function (ignored if kernel != 'poly')
}
svr = GridSearchCV(SVR(), param_grid, cv=5, n_jobs=-1, scoring='neg_mean_squared_error')
svr.fit(x_train,y_train)
svr_pred=svr.predict(x_test)
```

```python
svr.best_params_
```

```python
mse = mean_squared_error(y_test, svr_pred)
print("Mean Squared Error (MSE):", mse)
r2 = r2_score(y_test, svr_pred)
print("R Score:", r2)
mae = mean_absolute_error(y_test, svr_pred)
print("Mean Absolute Error (MAE):", mae)
```

```python
results_table1=pd.DataFrame([['SVM Regressor',mae,mse,r2]],columns=['Model','MAE','MSE','R']).round(3)
results_table=pd.concat([results_table,results_table1],ignore_index=True)
results_table
```

### 5. Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
param_grid = {
    'max_depth': [None, 10, 20, 30, 40, 50],               # Maximum depth of the tree
    'min_samples_split': [2, 5, 10, 20,40,50],                   # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4, 8,10,16],                      # Minimum number of samples required to be at a leaf node
    'max_features': [None, 'sqrt', 'log2'],                # Number of features to consider when looking for the best split
    'criterion': ['squared_error', 'absolute_error']        # Function to measure the quality of a split
}
dt=GridSearchCV(DecisionTreeRegressor(),param_grid=param_grid,n_jobs=-1,scoring='neg_mean_squared_error',cv=5)
dt.fit(x_train,y_train)
```

```python
dt.best_params_
```

```python
dt_pred=dt.predict(x_test)
```

```python
mse = mean_squared_error(y_test, dt_pred)
print("Mean Squared Error (MSE):", mse)
r2 = r2_score(y_test, dt_pred)
print("R Score:", r2)
mae = mean_absolute_error(y_test, dt_pred)
print("Mean Absolute Error (MAE):", mae)
```

```python
results_table1=pd.DataFrame([['Decision Tree',mae,mse,r2]],columns=['Model','MAE','MSE','R']).round(3)
results_table=pd.concat([results_table,results_table1],ignore_index=True)
results_table
```

### 6. Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
rf_pred=rf.predict(x_test)
```

```python
mse = mean_squared_error(y_test, rf_pred)
print("Mean Squared Error (MSE):", mse)
r2 = r2_score(y_test, rf_pred)
print("R Score:", r2)
mae = mean_absolute_error(y_test, rf_pred)
print("Mean Absolute Error (MAE):", mae)
```

```python
results_table1=pd.DataFrame([['Random Forest',mae,mse,r2]],columns=['Model','MAE','MSE','R']).round(3)
results_table=pd.concat([results_table,results_table1],ignore_index=True)
results_table
```

```python
results_table[['MAE','MSE','R']]=results_table[['MAE','MSE','R']].round(2)
```

##### Voting

```python
from sklearn.ensemble import VotingRegressor,AdaBoostRegressor,GradientBoostingRegressor
voting_reg=VotingRegressor(
    estimators=[
        ('EN',elas_net),
        ('SGD',sgd),
        ('RF',rf),
        ('DT',dt),
    ],
    weights=[3,3,2,1]
)
voting_reg.fit(x_train,y_train)
```

```python
voting_reg.score(x_test,y_test)
```

```python
vot_pred=voting_reg.predict(x_test)
```
